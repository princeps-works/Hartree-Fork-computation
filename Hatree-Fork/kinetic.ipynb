```python
import numpy as np
import Ipynb_importer
from scipy.sparse import diags
def kinetic_energy_with_Vext(l,r) -> np.ndarray : 
    """
    这部分计算的是动能部分，利用有限差分计算动能矩阵 T 

    参数:
    l 角动量量子数
    r 径向坐标列表
    
    输出:
    动能矩阵T
    """
    n = len(r)
    s = r[1] - r[0]

    # 二阶中心差分的三对角 (Dirichlet 边界)
    main = np.full(n, 1.0 / s**2, dtype=np.float64)
    off  = np.full(n-1, -0.5 / s**2, dtype=np.float64)

    # 径向势项 l(l+1)/(2 r^2)
    main += 0.5 * l * (l + 1) / np.maximum(r**2, 1e-14) # 防止 r=0 处除零

    # 组装三对角矩阵
    T = diags(diagonals=[off, main, off], offsets=[-1, 0, 1], format="csr")

    return T
#np.shape(kinetic_energy_with_Vext((1,0,2)))
```
